```python
class Sll_Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Sll_construction:
    def __init__(self):
        self.head = None
        
    def insertAtBegin(self, data):
        new_node = Sll_Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head = new_node
            
    def insertAtPosition(self, data, index):
        new_node = Sll_Node(data)
        if index == 0:
            self.insertAtBegin(data)
            return            # STOP further execution
        temp = self.head
        position = 0

        while temp is not None and position < index - 1:
            temp = temp.next
            position += 1
        if temp is None:
            print(f"Index {index} is out of bounds. Cannot insert.")
            return
        new_node.next = temp.next
        temp.next = new_node
            
    def removeFirstNode(self):
        if self.head is None:
            return "Empty List"
        else:
            removed_data = self.head.data     #set data of current head in removed_data ,30-->20-->10, removed_data = 30 to be deleted
            self.head = self.head.next        #Update the head to point to the next node, now pointer moves to 20
            return removed_data
            
    def removeParticularNode(self, index):
        if self.head is None:
            print("Empty List")
            return
        if index == 0:
            removed_data = self.head.data
            self.head = self.head.next  
            return
        temp = self.head
        position = 0
        
        while temp is not None and position < index - 1:
            temp = temp.next
            position += 1

        if temp is None or temp.next is None:
            print(f"Index {index} is out of bounds. Cannot delete.")
            return
    
        temp1 = temp.next 
        removed_data = temp1.data
        temp.next = temp1.next  
        temp1 = None 
        
    def traverseSLL(self):
        if self.head is None:
            print("The list is empty.")
        temp = self.head
        while temp is not None:
            print(temp.data, end=" --> ")
            temp = temp.next
        print(None)
        
    def SearchSLL(self,key):
        if self.head is None:
            print("The list is empty.")
        temp = self.head                                        #flag to track search
        position = 0
        while temp is not None:
            if temp.data == key:
                print(f"{key} is present in SLL at index number '{position}'")
                return
            temp = temp.next
            position+=1
        print(f"{key} is not present in SLL")
        
        
```

```python
def main():
    llist = Sll_construction()
    print("*** Linked List ***")
    while True:
        print("\n1. Insert At Beginning")
        print("2. Delete At Beginning")
        print("3. Show Entire List")
        print("4. Search a element in SLL")
        print("5. Insert At Particular Index")
        print("6. Delete At Particular Index")
        print("7. Exit")
        
        
        choice = int(input("Enter your choice: "))
        if choice == 1:
            value = eval(input("Enter the value to be inserted: "))
            llist.insertAtBegin(value)
            print(f"{value} inserted at the beginning.")
        elif choice == 2:
            removed_value = llist.removeFirstNode()
            if removed_value == "Empty List":
                print("Cannot delete. The list is empty.")
            else:
                print(f"Value Deleted: {removed_value}")
        elif choice == 3:
            llist.traverseSLL()
        elif choice == 4:
            key = eval(input("Enter the key you want to Search : "))
            llist.SearchSLL(key)
        elif choice == 5:
            value = eval(input("Enter the value to be inserted: "))
            index = int(input("Enter the index: "))
            llist.insertAtPosition(value,index)
            
        elif choice == 6:
            index = int(input("Enter the index: "))
            removed_value = llist.removeParticularNode(index)
            
        elif choice == 7:
            print("Exiting program. Goodbye! (❁´◡`❁) ")
            break
        else:
            print("Invalid Choice. Please try again.")
        
if __name__ == "__main__":
    main()
```
