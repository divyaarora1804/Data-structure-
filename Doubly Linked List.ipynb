```python
class dll_Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class dll_construction:
    def __init__(self):
        self.head = None
        
    def insertAtBegin(self, data):
        new_node = dll_Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
            
    def insertAtEnd(self,data):
        new_node = dll_Node(data)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next :
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp
        
    def removeFirstNode(self):
        if self.head is None:
            return "Empty List"
        else:
            removed_data = self.head.data     
            self.head = self.head.next        
            if self.head :
                self.head.prev = None
            return removed_data
        
    def removeEndNode(self):
        if self.head is None:
            return "Empty List"
        if self.head.next is None:
            removed_data = self.head.data     
            self.head = None 
            return removed_data
        temp = self.head
        while temp.next :
            temp = temp.next
        removed_data = temp.data
        temp.prev.next = None
        return removed_data
            
        
    def traverseDLL(self):
        if self.head is None:
            print("The list is empty.")
        temp = self.head
        while temp is not None:
            print(temp.data, end=" --> ")
            temp = temp.next
            
        print(None)
```

```python
def main():
    llist = dll_construction()
    print("*** Doubly Linked List ***")
    while True:
        print("\n1. Insert At Beginning")
        print("2. Delete At Beginning")
        print("3. Insert At End")
        print("4. Delete At End")
        print("5. Show Entire List")
        print("6. Exit")
        
        
        choice = int(input("Enter your choice: "))
        if choice == 1:
            value = eval(input("Enter the value to be inserted: "))
            llist.insertAtBegin(value)
            print(f"{value} inserted at the beginning.")
        elif choice == 2:
            removed_value = llist.removeFirstNode()
            if removed_value == "Empty List":
                print("Cannot delete. The list is empty.")
            else:
                print(f"Value Deleted: {removed_value}")
        elif choice == 3:
            value = eval(input("Enter the value to be inserted: "))
            llist.insertAtEnd(value)
            print(f"{value} inserted at the End.")
        elif choice == 4:
            removed_value = llist.removeEndNode()
            if removed_value == "Empty List":
                print("Cannot delete. The list is empty.")
            else:
                print(f"Value Deleted: {removed_value}")
        elif choice == 5:
            llist.traverseDLL()
        elif choice == 6:
            print("Exiting program. Goodbye! (❁´◡`❁) ")
            break
        else:
            print("Invalid Choice. Please try again.")
        
if __name__ == "__main__":
    main()
```
